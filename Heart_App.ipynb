```python
def my_app():
    import ipywidgets as widgets
    import ipywidgets
    



    def cust_plot(Disease):

        if Disease=="Heart Disease":
            print("Heart Disease Prediction")
            import ipywidgets
            file=open("C:/Users/DIGVIJAY/Downloads/pickle_folder/image/heart.png","rb")
            image=file.read()
                        
            a=ipywidgets.Image(value=image,format='jpg',width=320,height=400)
            display(a)
            
            import pickle
            
            import warnings
            warnings.filterwarnings('ignore')



            def heart(_BMI5CAT,_SMOKER3,CVDSTRK3,SEXVAR,CVDINFR4,_AGE_G,EXERANY2,
                         GENHLTH,CHCKDNY2,CHCOCNCR,CHCCOPD2,MARITAL,DIFFWALK):



                #importing the Randoom Forest Classifier Model


                with open('C:/Users/DIGVIJAY/Downloads/pickle_folder/random_forest_model.pkl','rb') as f:
                    new_rf=pickle.load(f)
                    

                _IMPRACE=4
                _PHYS14D=4

                input_dict={"BMI":_BMI5CAT,"SMOKING_HABBIT":_SMOKER3,"DIAGNOSED_WITH_STROKE":CVDSTRK3,"GENDER":SEXVAR
                            ,"EVER_DIAGNOSED_WITH_HEART_ATTACK":CVDINFR4,"AGE":_AGE_G,"RACE":_IMPRACE,
                            "EXERCISE_IN_PAST_30_DAYS":EXERANY2,"GENERAL_HEALTH":GENHLTH,"KIDNEY_DISEASE":CHCKDNY2,
                            "PHYSICAL_HEALTH_STATUS":_PHYS14D,"ANY_TYPE OF CANCER":CHCOCNCR,"CHRONIC_OPSTRUCTIVE_KIDNEY_DISEASE":CHCCOPD2,
                            "MARITAL_STATUS":MARITAL,"DIFFICULTY_IN_WALKING":DIFFWALK}                    
                    
                
                ####################   BMI    
                if(_BMI5CAT<1850):
                    _BMI5CAT=1
                elif(1850<=_BMI5CAT<2500):
                    _BMI5CAT=2
                elif(2500<=_BMI5CAT<3000):
                    _BMI5CAT=3
                else:
                    _BMI5CAT=4
                
                
                ############### SMOKER
                
                if(_SMOKER3=="Everday Smoker"):
                    _SMOKER3=1
                elif(_SMOKER3=="Someday Smoker"):
                    _SMOKER3=2
                    
                elif(_SMOKER3=="Former Smoker"):
                    _SMOKER3=3
                
                elif(_SMOKER3=="Non Smoker"):
                    _SMOKER3=4
                    
                else:
                    _SMOKER3=9
                    
                ######### Diagnosed With Stroke
                
                if(CVDSTRK3=="YES"):
                    CVDSTRK3=1
                elif(CVDSTRK3=="NO"):
                    CVDSTRK3=2
                elif(CVDSTRK3=="Refused"):
                    CVDSTRK3=9
                else:
                    CVDSTRK3=7
                    
                ######## Gender
                
                if(SEXVAR=="MALE"):
                    SEXVAR=1
                else:
                    SEXVAR=2
                    
                ######### Ever diagnosed with Heart Attack
                
                
                if(CVDINFR4=="YES"):
                    CVDINFR4=1
                elif(CVDINFR4=="NO"):
                    CVDINFR4=2
                elif(CVDINFR4=="Refused"):
                    CVDINFR4=9
                else:
                    CVDINFR4=7
                
                #### AGE
                _AGE_G
                
                if(18<=_AGE_G<=24):
                    _AGE_G=1
                elif(25<=_AGE_G<=34):
                    _AGE_G=2
                elif(35<=_AGE_G<=44):
                    _AGE_G=3
                elif(45<=_AGE_G<=54):
                    _AGE_G=4
                elif(55<=_AGE_G<=64):
                    _AGE_G=5
                else:
                    _AGE_G=6
                ##### RACE
                
                _IMPRACE=4
                
                ###### Exercise
                
                
                
                if(EXERANY2=="YES"):
                    EXERANY2=1
                elif(EXERANY2=="NO"):
                    EXERANY2=2
                elif(EXERANY2=="Refused"):
                    EXERANY2=9
                else:
                    EXERANY2=7
                
                
                ##### GENHEALTH
                if(GENHLTH=="EXCELLENT"):
                    GENHLTH=1
                    
                elif(GENHLTH=="VERY GOOD"):
                    GENHLTH=2
                elif(GENHLTH=="GOOD"):
                    GENHLTH=3
                elif(GENHLTH=="FAIR"):
                    GENHLTH=4
                elif(GENHLTH=="POOR"):
                    GENHLTH=5
                elif(GENHLTH=="Don't Know"):
                    GENHLTH=7
                else:
                    GENHLTH=9
                    
                ####### KIDNEY DISEASE
                
                if(CHCKDNY2=="YES"):
                    CHCKDNY2=1
                elif(CHCKDNY2=="NO"):
                    CHCKDNY2=2
                elif(CHCKDNY2=="Refused"):
                    CHCKDNY2=9
                else:
                    CHCKDNY2=7
                    
                ##### Physical Health Status
                _PHYS14D=4
                
                
                ##### Cancer
                
                
                if(CHCOCNCR=="YES"):
                    CHCOCNCR=1
                elif(CHCOCNCR=="NO"):
                    CHCOCNCR=2
                elif(CHCOCNCR=="Refused"):
                    CHCOCNCR=9
                else:
                    CHCOCNCR=7
                
                ######   Chronic Pulmonary Disease
                
                
                if(CHCCOPD2=="YES"):
                    CHCCOPD2=1
                elif(CHCCOPD2=="NO"):
                    CHCCOPD2=2
                elif(CHCCOPD2=="Refused"):
                    CHCCOPD2=9
                else:
                    CHCCOPD2=7
                    
                ###### Marital
                
           
                
                if(MARITAL=="MARRIED"):
                    MARITAL=1
                    
                elif(MARITAL=="DIVORCED"):
                    MARITAL=2
                elif(MARITAL=="WIDOWED"):
                    MARITAL=3
                elif(MARITAL=="SEPERATED"):
                    MARITAL=4
                elif(MARITAL=="NEVER MARRIED"):
                    MARITAL=5
                elif(MARITAL=="MEMBER OF UNMARRIED COUPLE"):
                    MARITAL=5
                else:
                    MARITAL=9
                    
                ### Diff
                
                
                if(DIFFWALK=="YES"):
                    DIFFWALK=1
                elif(DIFFWALK=="NO"):
                    DIFFWALK=2
                elif(DIFFWALK=="Refused"):
                    DIFFWALK=9
                else:
                    DIFFWALK=7


                test=[[_BMI5CAT,_SMOKER3,CVDSTRK3,SEXVAR,CVDINFR4,_AGE_G,_IMPRACE,EXERANY2,
                       GENHLTH,CHCKDNY2,_PHYS14D,CHCOCNCR,CHCCOPD2,MARITAL,DIFFWALK]]
                
                import ipywidgets
                def prediction_click(button):

                    y_pred=new_rf.predict(test)
                    if(y_pred==0):
                        
                        print("--------------------------------------------------------")
                        print("There is no chance of Heart Attack")
                        print("--------------------------------------------------------")
                    else:
                        
                        print("--------------------------------------------------------")
                        print("There is chance of Heart Attack")
                        print("--------------------------------------------------------")
                        
                        
                    def insert_database(button):
                        import pymongo
                        import ipywidgets
                        
                        client=pymongo.MongoClient("mongodb://localhost:27017")
                        db=client["HEART_DISEASE"]
                        HEART_INPUT=db["HEART_INPUT"]
                        db.HEART_INPUT.insert_one(input_dict)
                        print("THE VALUES OF INPUT INSERTED INTO DATABASE")

                       

                        

                    button=ipywidgets.Button(description="Click Here to Insert into Database")
                    button.on_click(insert_database)
                    display(button)


                button1=ipywidgets.Button(description="Click Here to Predict")
                button1.on_click(prediction_click)
                display(button1)



            import ipywidgets
            
            
            slider1 = ipywidgets.IntSlider(value=2175,min=0,max=9999,step=1,description='BMI :',
                                          disabled=False,continuous_update=True,orientation='horizontal')
            slider2 = widgets.RadioButtons(options=["Everday Smoker","Someday Smoker","Former Smoker","Non Smoker","Don't Know"]
                                           ,value='Non Smoker',layout={'width': 'max-content'},
                                           description='Smoking Habbit:',disabled=False)
            slider3 = widgets.RadioButtons(options=["YES","NO","Don't Know","Refused"]
                                           ,value='NO',layout={'width': 'max-content'},
                                           description='Diagnosed with stroke:',disabled=False)
            slider4 = widgets.RadioButtons(options=["MALE","FEMALE"],value='MALE',layout={'width': 'max-content'},
                                 description='Gender:',disabled=False)
            slider5= widgets.RadioButtons(options=["YES","NO","Don't Know","Refused"]
                                           ,value='NO',layout={'width': 'max-content'},
                                           description='Ever Diagnosed With Heart Attack:',disabled=False)
            slider6= ipywidgets.IntSlider(value=25,min=18,max=100,step=1,description='AGE :',
                                          disabled=False,continuous_update=True,orientation='horizontal')
            slider7 =  widgets.RadioButtons(options=["YES","NO","Don't Know","Refused"]
                                           ,value='NO',layout={'width': 'max-content'},
                                           description='Excersise in past 30 Days:',disabled=False)
            slider8 =  widgets.RadioButtons(options=["EXCELLENT","VERY GOOD","GOOD","FAIR",
                                                     "POOR","Don't Know","Refused"]
                                            ,value='GOOD',layout={'width': 'max-content'},
                                            description='General Health:',disabled=False)
            slider9 =  widgets.RadioButtons(options=["YES","NO","Don't Know","Refused"]
                                           ,value='NO',layout={'width': 'max-content'},
                                           description='Kidney Disease:',disabled=False)
            slider10 =  widgets.RadioButtons(options=["YES","NO","Don't Know","Refused"]
                                           ,value='NO',layout={'width': 'max-content'},
                                           description='Ever Had Cancer:',disabled=False)
            slider11 =  widgets.RadioButtons(options=["YES","NO","Don't Know","Refused"]
                                           ,value='NO',layout={'width': 'max-content'},
                                           description='Chronic Pulmonary Disease:',disabled=False)
            slider12 =  widgets.RadioButtons(options=["MARRIED","DIVORCED","WIDOWED","SEPERATED","NEVER MARRIED"
                                                      ,"MEMBER OF UNMARRIED COUPLE","REFUSED"]
                                           ,value='MARRIED',layout={'width': 'max-content'},
                                           description='Marital Status:',disabled=False)
            slider13 =  widgets.RadioButtons(options=["YES","NO","Don't Know","Refused"]
                                           ,value='NO',layout={'width': 'max-content'},
                                           description='Difficulty in Walking :',disabled=False)
            
            
            
            
            
            ipywidgets.interact(heart, _BMI5CAT=slider1,_SMOKER3=slider2,CVDSTRK3=slider3,SEXVAR=slider4,CVDINFR4=slider5
                                ,_AGE_G=slider6,EXERANY2=slider7,GENHLTH=slider8,CHCKDNY2=slider9
                                ,CHCOCNCR=slider10,CHCCOPD2=slider11,MARITAL=slider12,DIFFWALK=slider13)


            
    widgets.interact(cust_plot, Disease=["Select the Disease","Heart Disease"])
    
```
